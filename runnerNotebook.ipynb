```python
# %%
import pandas as pd
import easygui
import os
import subprocess

# %%

pathnames=easygui.fileopenbox("Please input file(s) to be run", "Multiple executer", filetypes= "*.txt", multiple=True)

```

```python


# %%
jz=easygui.buttonbox("Do you want to output the .gz file(s)?", choices=["Yes","No"])
print(jz)

# %%
configf=[]
for n, pathname in enumerate(pathnames):
    configf.append(os.path.basename(pathname))
    if n==0:
        configpath=(os.path.dirname(pathname))
    

    
print(configf)
gcsimpath=os.path.join(os.path.abspath(''))

# %%
```

```python
char = "total"
DPSlist=[]
if (jz=="Yes"):
    print("Output is .gz")
    if(char=="total"):
        for f, configs in enumerate(configf):
            stream = subprocess.run([gcsimpath+"\gcsim.exe", '-c=' + configpath+"\\"+configf[f],"-gz","-out="+configpath+"\\"+str(configf[f]).replace(".txt","")], capture_output=True)
            strStream=str(stream.stdout).split("\\n")

            tokens = strStream
            #print(tokens)
            startidx = 0
            endidx = 0
            iDPS=0
            for i, token in enumerate(tokens):
                if token.split(" ")[0] == "Damage":
                    print(token)
                    if len(token.split(" ")) == 7:
                        #print("OMMITED")
                        if startidx == 0:
                            startidx = i+2
                            endidx = i
                            iDPS=8
                        else:
                            endidx = i
                            break
                    else:
                        if startidx == 0:
                            startidx = i+17
                            endidx = i
                            iDPS=5
                        else:
                            endidx = i
                            break
            totalvalues=str(tokens[startidx]).split()
            #print(totalvalues)
            actDPS=totalvalues[iDPS]
            DPSlist.append(actDPS.replace(';',''))
            print(configf[f]+":     "+actDPS)


    else:
        for f, configs in enumerate(configf):

            stream = subprocess.run([gcsimpath+"\gcsim.exe", '-c=' + configpath+"\\"+configf[f],"-gz","-out="+configpath+"\\"+str(configf[f]).replace(".txt","")], capture_output=True)
            strStream=str(stream.stdout).split("\\n")
            
            #print(strStream)

            tokens = strStream
            startidx = 0
            endidx = 0
            for i, token in enumerate(tokens):
                if token.split(" ")[0] == char:
                    if startidx == 0:
                        startidx = i
                    else:
                        endidx = i
                        break
            yaevalues=str(tokens[endidx]).split()
            print(yaevalues)
            actDPS=yaevalues[4]
            DPSlist.append(actDPS.replace(';',''))
            
            print(configf[f]+":     "+actDPS)

else:
    print("No .gz output")
    if(char=="total"):
        for f, configs in enumerate(configf):
            stream = subprocess.run([gcsimpath+"\gcsim.exe", '-c=' + configpath+"\\"+configf[f]], capture_output=True)
            strStream=str(stream.stdout).split("\\n")

            tokens = strStream
            #print(tokens)
            startidx = 0
            endidx = 0
            iDPS=0
            for i, token in enumerate(tokens):
                if token.split(" ")[0] == "Damage":
                    print(token)
                    if len(token.split(" ")) == 7:
                        #print("OMMITED")
                        if startidx == 0:
                            startidx = i+2
                            endidx = i
                            iDPS=8
                        else:
                            endidx = i
                            break
                    else:
                        if startidx == 0:
                            startidx = i+17
                            endidx = i
                            iDPS=5
                        else:
                            endidx = i
                            break
            totalvalues=str(tokens[startidx]).split()
            print(totalvalues)
            actDPS=totalvalues[iDPS]
            DPSlist.append(actDPS.replace(';',''))
            print(configf[f]+":     "+actDPS)


    else:
        for f, configs in enumerate(configf):

            stream = subprocess.run([gcsimpath+"\gcsim.exe", '-c=' + configpath+"\\"+configf[f]], capture_output=True)
            strStream=str(stream.stdout).split("\\n")
            
            #print(strStream)

            tokens = strStream
            startidx = 0
            endidx = 0
            for i, token in enumerate(tokens):
                if token.split(" ")[0] == char:
                    if startidx == 0:
                        startidx = i
                    else:
                        endidx = i
                        break
            yaevalues=str(tokens[endidx]).split()
            print(yaevalues)
            actDPS=yaevalues[4]
            DPSlist.append(actDPS.replace(';',''))
            
            print(configf[f]+":     "+actDPS)


# %%
```

```python
print(DPSlist)

# %%
```

```python

df = pd.DataFrame({'Name' : configf, 'DPS' : DPSlist})    
df['Name']=df['Name'].str.replace('.txt','',regex=True)#Delete .txt 
df['Name']=df['Name'].str.replace('_',' ')#Replace _s with spaces 
df['DPS']=pd.to_numeric(df['DPS'])#convert to numbers
df=df.sort_values(by='DPS',ascending=False)#sort
df['%']=df['DPS']/df['DPS'].median()*100#create a new column with % of the median
df.index=df['Name']
del df['Name']

#df.to_csv('items.csv', index=False, encoding='utf-8')
print(df)

exportExcel=input("Do you want to export the results to Excel? (enter Y for Yes): ")
if exportExcel=="Y":
    df.to_excel("Results.xlsx")
    print("Exported to Results.xlsx")
```
